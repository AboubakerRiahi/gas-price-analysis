```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
gas=pd.read_csv('gas_prices.csv')
gas
```

**gaz prince USA in years**

```python
plt.figure(figsize=(10,5))
plt.plot(gas['Year'],gas['USA'],'r+--')
```

**showing gas prices over the years for multiple countries**

```python
plt.figure(figsize=(10,5))

#plt.plot(gas['Year'],gas['USA'],'b.-')
for country in gas.columns[1::]:
          plt.plot(gas['Year'],gas[country],marker='o',label=country)

print(gas['Year'].tolist())
print(gas['Year'][::3].tolist()) #[1990, 1993, 1996, 1999, 2002, 2005, 2008]
plt.xticks(gas['Year'][::3].tolist())
plt.title('Gas Prices (in USD)',fontdict={'fontweight':'bold','fontsize':15})
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()
plt.savefig('gas_price_figure.png',dpi=300)
plt.show()
```
